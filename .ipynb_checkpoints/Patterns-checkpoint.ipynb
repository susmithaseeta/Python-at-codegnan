```python
n = 5
for i in range(n):
    for j in range(n):
        if i == j or j == n - 1:
            print("*", end = " ")
        else:
            print("*", end = " ")
    print()
```

```python
n = 5
for i in range(n):
    for j in range(n):
        if i == j or j == n - i - 1:
            print("*", end = " ")
        else:
            print("+", end = " ")
    print()
```

```python
n = 5
k = 1
for i in range(n):
    for j in range(n):
        if j >= n - i - 1:
            print(k,end = "\t")
            k += 1
        else:
            print(" ", end = "\t")
    print()
```

```python
# Hallow square
n = 5
for i in range(n):
    for j in range(n):
        if i == 0 or j == 0 or i == n - 1 or j == n - 1:
            print('*',end = " ")
        else:
            print(' ',end = " ")
    print()
```

```python
# Hallow lower right angle triangle
n = 5
for i in range(n):
    for j in range(n):
        if i == j or  i == n - 1 or j == 0:
            print('*',end = " ")
        else:
            print(' ',end = " ")
    print()
```

```python
n = 5

for i in range(n):
    for j in range(n):
        if i %2 == 0:
            print(i * n + j + 1,end = "\t")
            
        else:
            print((i + 1) * n - j, end = "\t")
            #print(" ", end = "\t")
    print()
```

```python
n = 5
k = n * (n+1)//2
for i in range(n):
    for j in range(n):
        if i>=j:
            print(k,end = "\t")
            k-=1
        else:
            print(" ", end = "\t")
    print()
```
