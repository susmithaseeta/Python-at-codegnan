### Set Characteristics

```python
# Set unordered collection
# Set return unique elements
s = {1,2,3,2,3,'Hi','a',2.02, (10,20,30)}
print(s)
# print(s[0]) # set does not support indexing
```

```python
# List can't be used as set element
s = {1,2,[3,2,3],'Hi','a',2.02, (10,20,30)}
print(s)
```

### Set Methods

```python
# Element adding method 
# add(ele): It adds the elements to set if element not in set, otherwise nothing
s = {1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add("Code")
print(s)
```

- Element removing methods
  - remove(item): It removes the item if it exists in set otherwise 'KeyError'
  - discard(item): It removes the item if it exists in set otherwise nothing

```python
s = {1,2,3,4,5}
s.remove(5)
print(s)
# s.remove(100) # KeyError
s.discard(3)
print(s)
s.discard(100) 
print(s)
```

## Sets Operations
  - Union(set): It returns the all unique elements all sets.( | )
  - Intersection(set): It return common elements from all sets.( & )
  - Difference(set): It returns the a set elemenrs those are not in another sets.( - )
  - symmetric_difference(set) It return the non common elements from all sets.( ^ )

```python
# union(set)
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,8,9}
union = s1.union(s2)
print(union)
union = s1.union(s2,s3)
print(union)
res = s1 | s2 | s3
print(res)
```

```python
# intersection(set)
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,7,8,9}
intersection = s1.intersection(s2)
print(intersection)
intersection = s1.intersection(s2,s3)
print(intersection)
res = s1 & s2 & s3 # common values
print(res)
```

```python
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,7,8,9}
s1.intersection_update(s2,s3)
print(s1,s2,s3)
```

```python
# difference(set)
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,7,8,9}
difference = s1.difference(s2)
print(difference)
difference = s1.difference(s2,s3)
print(difference)
res = s1 - s2 
print(res)
print((s1 | s2) - s3)
```

```python
# symmetric_difference(set)
s1 = {1,3,5,7}
s2 = {4,5,6,7}
s3 = {1,2,7,8,9}
sym_diff = s1.symmetric_difference(s2)
print(sym_diff)
res = s1 ^ s2 ^s3
print(res)
```

```python
# issubset(set)
# issuper(set)
# isdisjoin(set)
s1 = {1,3,5,7}
s2 = {1,3,4,5,6,7}
print(s1.issubset(s2))
print(s2.issubset(s1))
print(s2.issuperset(s1))
print(s2.isdisjoint(s1))
```
