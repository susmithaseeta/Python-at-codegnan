```python
# characteristics
# ordered collection
t = (1,2,3,[10,30,40],23)
print(t)
print(t[3][2])
```

```python
#immutable
# After operation it doesn't change the length of the object and dosen't change remaining elemnts position
t = (1,2,3,[10,30,40],23)
t[3].append(100)
print(t)
```
