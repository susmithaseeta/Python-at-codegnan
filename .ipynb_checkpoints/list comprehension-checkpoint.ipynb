```python
#generate 1-n numbers using list comprehension
n = int(input())
res = [x for x in range(1,n+1)]
print(res)
```

```python
# fist n even numbers using list comprehension
n = int(input())
res = [x*2 for x in range(0, n)]
print(res)
```

```python
# fist the first n quibic number
n = int(input())
res = [x**3 for x in range(1 , n + 1)]
print(res)
```

```python
# find off numbers between n1 to n2
n1,n2 = map(int,input().split())
res = [x for x in range(n1 , n2 + 1, 2)] # print even 
print(res)
```

```python
# find off numbers between n1 to n2
n1,n2 = map(int,input().split())
res = [x for x in range(n1 , n2 + 1,) if x % 2 == 1] # print odd
print(res)
```

```python
# 2 dices combination values
res = [(x,y) for x in range (1,7) for y in range(1,7)]
print(res)
```

```python
# return the dice combination if sum of two face value is div by 3
res = [(x,y) for x in range (1,7) for y in range(1,7) if (x +y) % 3 == 0]
print(res)
```

```python
# print any dice whose value is 4 
res = [(x,y) for x in range (1,7) for y in range(1,7) if x == 4 or y == 4]
print(res)
```

```python
# count the number of vowels exists in string 
s = input()
res = [char for char in s.lower() if char in "aeiou" ]
print(res)
print(len(res))
```

```python
# or 
s = input()
res = sum ([1 for char in s.lower() if char in "aeiou" ])
print(res)
```

```python
# set comprehension 
# 1 to n numbers
n = int(input())
res = {x for x in range (1,n+1)}
print(type(res))
print(res)
```

```python
# dictionary comprehension 
# 1 to n numbers
n = int(input())
res = {x : x ** 2 for x in range (1,n+1)}
print(type(res))
print(res)
```

```python
# tuple comprehension 
# 1 to n numbers
n = int(input())
res = tuple(x for x in range (1,n+1))
print(type(res))
print(res)
```

```python
# weather the person is eligible for vote or not
age = int(input())
if age >= 18:
    print("Eligiable")
else:
    print("Not Eligiable")
```

- Assert Keyword : This keyword is used for testing 

```python
age = int(input())
assert age >= 18, " Age must be above 17"
print("Eligiable")
```
