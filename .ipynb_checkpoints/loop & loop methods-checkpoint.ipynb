```python
# check which one is even or odd without loop
a,b,c = map(int,input().split())
if a % 2 == 0:
    print(a," is even")
else:
    print(a," is odd")
if b% 2 == 0:
    print(b," is even")
else:
    print(b," is odd")
if c% 2 == 0:
    print(c," is even")
else:
    print(c," is odd")
```

```python
#check which one is even and which and is odd
lst = [10,12,13]
for num in lst:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
# even in exponent n**2, odd (n+1)
lst = list(map(int,input().split()))
for num in lst:
    if num % 2 == 0:
        print(num**2)
    else:
        print(num+1)
```

```python
lst = list(map(int,input().split()))
for num in lst:
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 5 == 0:
        print("Buzz")
    elif num % 3 == 0:
        print("Fizz")
    else:
        print(f"{num}")     
```

```python
lst = list(map(int,input().split()))
res = []
for num in lst:
    if num % 3 == 0 and num % 5 == 0:
        res.append("FizzBuzz")
    elif num % 5 == 0:
        res.append("Buzz")
    elif num % 3 == 0:
        res.append("Fizz")
    else:
        res.append(f"{num}")  
print(res)
```

```python
# count vowels and consonents
s = input().lower()
c = v = 0
for char in s:
    if char.isalpha():
        if char in 'aeiou':
            v += 1
        else:
            c += 1
print(f"Vowels: {v}, Consonants: {c}")
```

```python
# print 1 to n number
n = int(input())
for i in range(1, n+1):
    print(i, end = " ")
```

```python
# print 1 to 10 in while loop
n = int(input())
i = 1
while i <= n:
    print(i,end=" ")
    i += 1
```

```python
lst = [11, 12, 9, 5, 6, 7, 13]
target = 5
n = len(lst)
for i in range(n):
    if lst[i] == target:
        print(i)
        break
else:
    print(-1)
```

```python
# Prime Number
n = int(input())
factors = list()
for i in range(1,n+1):
    if n % i == 0:
        factors.append(i)
print(factors)
```

```python
n = int(input())

if n <= 1:
    print(f"{n} is not a prime number")
else:
    is_prime = True
    for i in range(2, n):  
        if n % i == 0:    
            is_prime = False
            break
    if is_prime:
        print(f"{n} is a prime number")
    else:
        print(f"{n} is not a prime number")
```

```python
# Prime number
n = int(input())
for i in range (2,n//2+1):
    if n % i == 0:
        print(f"{n} not prime")
        break
else:
    print(f"{n} is prime")
```

```python
# Prime number
n = int(input())
for n in range(2,n):
    for i in range (2,n//2+1):
        if n % i == 0:
            print(f"{n} not prime")
            break
else:
    print(f"{n} is prime")
```
