```python
# check which one is even or odd without loop
a,b,c = map(int,input().split())
if a % 2 == 0:
    print(a," is even")
else:
    print(a," is odd")
if b% 2 == 0:
    print(b," is even")
else:
    print(b," is odd")
if c% 2 == 0:
    print(c," is even")
else:
    print(c," is odd")
```

```python
#check which one is even and which and is odd
lst = [10,12,13]
for num in lst:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
# even in exponent n**2, odd (n+1)
lst = list(map(int,input().split()))
for num in lst:
    if num % 2 == 0:
        print(num**2)
    else:
        print(num+1)
```

```python
lst = list(map(int,input().split()))
for num in lst:
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 5 == 0:
        print("Buzz")
    elif num % 3 == 0:
        print("Fizz")
    else:
        print(f"{num}")     
```

```python
lst = list(map(int,input().split()))
res = []
for num in lst:
    if num % 3 == 0 and num % 5 == 0:
        res.append("FizzBuzz")
    elif num % 5 == 0:
        res.append("Buzz")
    elif num % 3 == 0:
        res.append("Fizz")
    else:
        res.append(f"{num}")  
print(res)
```

```python
# count vowels and consonents
s = input().lower()
c = v = 0
for char in s:
    if char.isalpha():
        if char in 'aeiou':
            v += 1
        else:
            c += 1
print(f"Vowels: {v}, Consonants: {c}")
```

```python
# print 1 to n number
n = int(input())
for i in range(1, n+1):
    print(i, end = " ")
```
