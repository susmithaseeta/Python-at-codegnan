## Scopes of variables
 - 1. Local Scope
 - 2. Enclosing Scope
 - 3. Global Scope
 - 4. Build-in Scope

```python
x = 10 # Global Scope Variable
def outer():
    y = 20 # Enclosing Scope Variable
    def inner():
        z = 30 # 1local Scope Variable
        print(z,y,x) # print() - Build-in Scope function
    inner()
outer()
```

```python
#using global scope variable in local scope
x = 100
def outer():
    y = x + 1
    print("Local Scope variable value:",y)
outer()
print("Global scope value:",x) 
```

```python
# Updating the  global scope variable value in local scope by using ' global ' keyword
x = 100
def outer():
    global x
    x = x + 1
    print("Local Scope variable value:",x)
outer()
print("Global scope value:",x)
```

```python
# Updating the Enclosing Scope variable in local scope
x = 10 # global scope variable
def outer():
    y = 20 # Enclosing Scope Variable
    def inner():
        z = 30 # local Scope Variable
        y = y + z
        print(z,y,x) # print() - Build-in Scope function
    inner()
outer()
```

```python
# Updating the Enclosing Scope variable in local scopeby using non-local
x = 10 # global scope variable
def outer():
    y = 20 # Enclosing Scope Variable
    def inner():
        nonlocal y
        z = 30 # local Scope Variable
        y = y + z
        print(z,y,x) # print() - Build-in Scope function
    inner()
    print("Enclosing scxope variable value:",y)
outer()
```

```python
x = 100 # global scope variable
def outer():
    x = 200
    x = x + 2
    print("Local scope variable x value:",x)
outer()
print("Global scope variable x value:",x)
```
