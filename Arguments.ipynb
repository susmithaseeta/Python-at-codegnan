```python
# Type of Arguments
# 1. Positional Arguments : Here arguments are assigns to parameter as per the sequential 
# function defination block
def sub(x:int,y:int)->int:
    z = x - y
    return z
# function calling block
a, b = 20, 10
print(sub(a, b))
print(sub(b, a))
```

```python
# 2. Key word arguments: Arguments are explicity passed with parameter names
# function defination block
def sub(x:int,y:int)->int:
    z = x - y
    return z
# function calling block
a, b = 20, 10
print(sub(x = a,y = b))
print(sub(y = b,x = a))
```

```python
# Addition operation
def add(x:int,y:int)->int:
    z = x + y
    return z
# function calling block
a, b = 20, 10
print(add(x = a,y = b))
print(add(y = b,x = a))
print(add(5, y = 10))
```

```python
#  3-variable addition
# function defination block
def add(x:int,y:int,z:int)->int:
    p = x + y + z
    return p
# function calling block
a, b ,c= map(int,input().split())
print(add(a,b,c))
```

```python
# 4-variable addition
# function defination block
def add(w:int,x:int,y:int,z:int)->int:
    p = w + x + y + z
    return p
# function calling block
a, b ,c, d= map(int,input().split())
print(add(a,b,c,d))
```

```python
# Default Arguments: The default arguments are optional arguments, if you pass value for default argument,
# in that case it assign default avlue otherwise, it assign new arguments value. 
# function defination block 
def add(w:int,x:int,y:int = 0,z:int = 0)->int:
    p = w + x + y + z
    return p
# function calling block
a, b ,c, d= map(int,input().split())
print(add(a,b))
print(add(a,b,c))
print(add(a,b,c,d))
```

```python
# multiplication operation
# function defination block
def mul(w:int,x:int,y:int = 1,z:int = 1)->int:
    p = w * x * y * z
    return p
# function calling block
a, b ,c, d= map(int,input().split())
print(mul(a,b))
print(mul(a,b,c))
print(mul(a,b,c,d))
```

```python
# Variable Length Positional arguments
def add(x,y = 0, *args):

    print(type(args))
    print(args)
    s = 0
    for val in args:
        s+= val
    return s
print(add(10,20,1,2,3,4,5))
```

```python
# 5. Variable length Keyword Arguments
def add(x, y = 0, *args, **kwargs):
    print(type(kwargs))
    print(kwargs)
    print(kwargs.values())
    print(kwargs.keys())

print(add(1,2,3,4,5, name = "Susmitha", age = 20, course = "PFS"))
```

```python
# In Python every type function caaling block shares the arguments reference to parameters
#  if the arguments are immutable data types, then it behaves likes pass by value and if 
#  arguments are mutable data types(i.e list, set, dict) then it behaves likes pass by reference

# Pass by value:
def fun(x):
    print(id(x))
    x += 2
    print("After add 2")
    print(id(x))
    print(x)
a = 10
print("Before Function Execution 'a' value:",a)
fun(a)
print("After Function Execution 'a' value:",a)
print(id(a))
print(a)
```

```python
# Pass by Reference
def add_ele(lst, ele):
    lst.append(ele)

arr = [1,2,3,4]
print("Before Function execution arr value", arr)
add_ele(arr, 100)
print("After Function execution arr value", arr)
```

```python
# Pass by Reference
def add_ele(lst, ele):
    lst=[10,20,30]

arr = [1,2,3,4]
print("Before Function execution arr value", arr)
add_ele(arr, 100)
print("After Function execution arr value", arr)
```
