```python
# find the difference between two string 
s1 = "CODE"
s2 = "GNAN"
n = len(s1)
s = ""
for i in range(n):
    diff = abs(ord(s1[i])-ord(s2[i]))
    if diff == 0:
        s += s1[i]
    else:
        s += chr(64+diff)
print(s)
```

```python
# find bigger and smaller character from s1 and s2
# compare each character in both strins and if s1[i] > s2[i] to bigger and s2[i] to smaller 
# if both same leave it
s1, s2 = map(str,input().upper().split())
n = len(s1)
big = small = ""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i] > s2[i]:
            big += s1[i]
            small +=s2[i]
        else:
            big +=s2[i]
            small +=s1[i]
print(big)
print(small)
```

```python
# map(fun,iterable) : It applies the functionalty on each element in iterable
# fulter(fun, iterable) : It filters the itables, it returns the values which is not None
# find the squere of list of integers
def squere(x):
    return x**2
lst = [1,2,3,4,5]
print(list(map(squere, lst)))
```

```python
# filter the even numbers from list of integers 
def even(x):
    if x % 2 == 0:
        return x
        
arr = [1,2,3,4,5,6,7]
print(filter(even,arr))
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
def squere(t):
    x, y = t
    return x + y
lst = [(1,2),(3,4)]
print(list(map(squere,lst)))
```

```python
# enumerate(itable): It returns the list of tuple of index & value
arr = [10,20,30]
print(enumerate(arr))
print(list(enumerate(arr)))
```

```python
# find the target element postion in list
arr = [10, 20, 30, 40, 50]
target = 40
for ind, val in enumerate(arr):
    if target == val:
        print(ind)
        break
else:
    print(-1)
```

```python
# any(iterable): It returns True if any one iterable element's boolean value is True
arr = [1,2,4,0,0,0]
print(any(arr))
print(any([0,0,0,0]))

# all(iterable): It returns True when all  iterable element's boolean value is True
print(all([1,2,3,4]))
print(all([True, False, True]))
print(all([False, False, False]))
```
