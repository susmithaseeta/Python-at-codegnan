```python
# Check weather the product stock is available or not (read product name and stock)
product = input() # string 
stock = int(input()) # integer
if stock > 0:
    print(f"currently {product} : {stock} item available")
else:
    print(f" currently {product} out of stock")
```

```python
#  In a shop contains some items along with price like {'pen':5, "book":20, "pencil":3}. A customer want some item from shop,if item is available 
# in shop, then find the total amount, otherwise 'Item not available'
products = {'pen':5, 'book':20, 'pencil':3}
item = input().lower()
quantity = int(input())
if item in products: # (in: membership operator)
    total_amount = products.get(item) * quantity
    print(f"Total amount of {quantity} {item} is: {total_amount}")
else:
    print(f"{item} not available")
    
```
