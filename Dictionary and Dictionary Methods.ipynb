### Dictonary Characteristics


```python
# Dictionary ordered collection
details = {"Name" : "seeta", "Course" : "python", "Batch" : 57, "Age" : 21}
print(details)
```

```python
# Dictionary values access by using keys
# Updating dictionary values
# Adding new items to dictionary
details = {"Name" : "seeta", "Course" : "python", "Batch" : 57,"Age" : 21}
print(details['Course'])
# print(details['location']) # KeyError : 'Location'
details['Age']=22
print(details['Age'])
# Adding new items
details['Location'] = "Hyderabad"
print(details)
```

```python
# adding age 
details = {"Name" : "seeta", "Course" : "python", "Batch" : 57,"Age" : 21}
details['Age'] = details['Age'] + 1
print(details)
```

```python
# dictionary keys are unique
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57, 'Age': 22, 'Location': 'Hyderabad'}
print(details)
```

```python
# List can't be used as dictionary key
# dictionary keys are unique
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', [1,2,3]: 57, 'Age': 22, 'Location': 'Hyderabad'}
print(details)
```

```python
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57, 'Age': 22, 'Location': 'Hyderabad', 'Marks' : [65,54,90]}
print(details)
```

### Dictionary Methods


- 'get(key, default_value:optional)' : It returns the value if key is exists in dictionary otherwise it return default value. 

```python
# dictionary keys are unique
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57, 'Age': 22, 'Location': 'Hyderabad'}
print(details['Name'])
# print(details['marks']) # This one raises KeyError
print(details.get('Name'))
print(details.get('Marks')) # get Method by-default it return None if key not exists in dictionary
print(details.get('Marks', "Marks not avaliable")) # it returns default values because, 
```

### Update Method
- 'update(new_dictionary)': This method updates the values if new_dictionary keys already exists in Dictionary , otherwise it adds new items to dictionary 


```python
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57}
details.update({'Batch': 58, 'Course': 'Java', 'Location': 'Hyd'})
print(details)
```

- 'keys()': This method returns the list dictionary keys
- 'values()': This method returns the list dictionary Values
- 'items()': This method returns the list of Tuple of key-value pairs

```python
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57}
print(details.keys())
print(details.values())
print(details.items())
```

```python
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57}
keys=details.keys()
print(type(keys))
print(keys)
```

### Item removing Methods 

- popitem(): It return and removes the last item as tuple.
- pop(key): This method removes the specified key item and return the specified key's value, otherwise 'keyError'.

```python
details = {'Age' : 23, 'Name': 'seeta', 'Course': 'python', 'Batch': 57, 'Location': 'Hyd'}
item = details.popitem()
print(item)
print(details)
val = details.pop('Name')
print(val)
print(details)
```
