## Input Formating 

```python
#reading integer from user
num = int(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# Converting or reading float value from user
val = float(input())
print(type(val))
print(val)
```

```python
# Complex reading from user
val = complex(int(input()), int(input()))   # it execute left to right
print(type(val))
print(val)
```

```python
# Boolean reading
val = bool(input()) 
print(type(val))
print(val)
```

```python
# Boolean reading as integer
val = bool(int(input())) 
print(type(val))
print(val)
```

```python
# reading string value from user
s = input()
print(type(s))
print(s)
```

```python
# reading list of integers from user
lst = input()
print(lst)
```

```python
lst = input().split()
print(lst)
```

```python
lst = map(int, input().split())
print(lst)
```

```python
lst = list(map(int, input().split()))
print(lst)
```

```python
# reading list of string from user
s = input().split()
print(s)
```

```python
s = list(map(str, input().split()))
print(s)
```

```python
lst = input().split()
print(lst)
```

```python
# reading Tuple of int from user
lst = tuple(map(int, input().split()))
print(lst)
```

```python
lst = input().split()
print(tuple(lst))
```

```python
# reading set of int from user
lst = set(map(int, input().split()))
print(lst)
```

```python
# eval(): This function evaluates the string type expression
val = eval(input())
print(type(val))
print(val)
```

## Output Formatting

```python
name = input()
age = int(input())
print("My name is", name, "and I am" ,age, "years old")
```

```python
#using string concatination
name = 'susmitha'
age = 21
print("My name is " + name + " and I am " + str(age) + " years old")
```

```python
# 1. string formatting using comma (old style formate)
name = 'susmitha'
age = 21
print("My name is", name, "and I am" ,age, "years old")
```

```python
# 2. Access specifiers
# string - %s
# integer - %d
# float - %f
name = 'susmitha'
age = 21
# my name is susmitha and i am 21 years old
print("My name is %s and I am %d years old"%(name,age))
```

```python
name = 'susmitha'
age = 21
# my name is susmitha and i am 21 years old
print("My name is %s and I am %d years old"%(age,name))
```

```python
# 3. Dot Format
name = 'susmitha'
age = 21
# my name is susmitha and i am 21 years old
print("My name is {} and I am {} years old".format(name,age))
print("My name is {} and I am {} years old".format(age,name))
```

```python
# 4. f-Format
name = 'susmitha'
age = 21
# my name is susmitha and i am 21 years old
print("My name is {name} and I am {age} years old")
print("My name is {age} and I am {name} years old")
```
