### Input Formating 

```python
#reading integer from user
num = int(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# Converting or reading float value from user
val = float(input())
print(type(val))
print(val)
```

```python
# Complex reading from user
val = complex(int(input()), int(input()))   # it execute left to right
print(type(val))
print(val)
```

```python
# Boolean reading
val = bool(input()) 
print(type(val))
print(val)
```

```python
# Boolean reading as integer
val = bool(int(input())) 
print(type(val))
print(val)
```

```python
# reading string value from user
s = input()
print(type(s))
print(s)
```

```python
# reading list of integers from user
lst = input()
print(lst)
```

```python
lst = input().split()
print(lst)
```

```python
lst = map(int, input().split())
print(lst)
```

```python
lst = list(map(int, input().split()))
print(lst)
```

```python
# reading list of string from user
s = input().split()
print(s)
```

```python
s = list(map(str, input().split()))
print(s)
```

```python
lst = input().split()
print(lst)
```

```python
# reading Tuple of int from user
lst = tuple(map(int, input().split()))
print(lst)
```

```python
lst = input().split()
print(tuple(lst))
```

```python
# reading set of int from user
lst = set(map(int, input().split()))
print(lst)
```

```python
# eval(): This function evaluates the string type expression
val = eval(input())
print(type(val))
print(val)
```
