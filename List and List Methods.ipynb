## List Characteristcs

```python
# Define List
# List is ordered Collection
lst = [1,2,4.5,"Hi","Hello",1,3,4]
print(lst)
```

```python
# List allows indexing and Slicing
# Accessing List Item
lst = [1,2,4.5,"Hi","Hello",1,3,4]
print(lst)
print(lst[4])
print(lst[-3])
print(lst[::4])
print(lst[3:4])
```

```python
lst = [10,20,30,40,50,60,70]
print(lst)
print(lst[-4:10])
print(lst[:5:-1])
print(lst[:-5:-1])
print(lst[:500])
print(lst[::-1]) # reverse list
```

```python
# Empty List Creation
l1 = []
l2 = list()
print(type(l1),type(l2))
```

```python
# List Opersations
# 1. List Concatination ( using + operator)
# 2. List Repeatation ( using * operator)
l1 = [1,2,3]
l2 = [3,4,5,6]
print(l1+l2)
print(l1+l1)
print(l1 * 3)
print([0] * 10)
```

### List Methods
- Item adding methods
  - append(item): It adds item at end of list.
  - extend(iterable): It adds the all iterator elements at end of list.
  - insert(pos, item): It adds item at specified position.

```python
lst=[10,20,30,40]
lst.append(100) # Append Method
print(lst)
lst.extend([1,2,3]) # Extend Method
print(lst)
lst.append([100,200]) 
print(lst)
lst.insert(0,1000)
print(lst)
```

- Item Removing Methods=>
   - remove(item): It Removes the first occurance item from the list if it is exists in list otherwise 'ValueError'.
   - pop(pos): It remove and return the specific position item if position is passed, otherwise it remove and returns the last item from list.
   - clear(): It removes all elements from list

```python
lst = [10,20,30,40]
print(lst)
print(lst.remove(20))
print(lst)
```

```python
lst = [10,20,30,40,50,20]
print(lst.remove(20))
print(lst)
# print(lst.remove(100))
print(lst)
```

```python
lst = [10,20,30,40,50,20]
val = lst.pop()
print(val,lst)
print(lst)
val2 = lst.pop(0)
print(val2)
print(lst)
lst.clear() #clear method
print(lst)
```

- index and count()
  - index(item): If item is present in list it returns the index otherwise 'ValueError'
  - count(item): It return the count of Total occurances of item


```python
lst = [10,20,30,20,40,50]
ind = lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# Sorting Elements / Items
# sort(reverse = False): By default it sort the origin list in into Ascending order.
#                           if reverser is True it sort in decending order.
lst = [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```

```python
lst = ['hi','hello','banti']
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```

```python
# Build-in FUnction- Len(), min(), max(), sum()
lst=[1,2,3,4,5]
print(len(lst),min(lst),max(lst),sum(lst))
print("Average:",sum(lst) / len(lst))
```

```python
lst = ["a","A","Bhavani","Seeta","Luna","kalix"]
print(len(lst),min(lst),max(lst))
```
