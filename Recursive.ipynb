```python
# factorial of number
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

```python
# sum of digits in a number
# n = 456, o/p = 15
def sumOfDigits(n):
    if n == 0:
        return 0
    return n % 10 + sumOfDigits(n//10)
print(sumOfDigits(456))
print(sumOfDigits(15346))
```

```python
# count digits in a number
# n = 456, o/p = 3
def sumOfDigits(n):
    if n == 0:
        return 0
    return 1 + sumOfDigits(n//10)
print(sumOfDigits(456))
print(sumOfDigits(15346))
```

```python
# find  the squere of a number using lambda
s = lambda x : x ** 2
print(s(10))
```

```python
# find the squeres of list of integers using map & lambda
arr = [1,2,3,4,5]
print(list(map(lambda x : x ** 2, arr)))
```

```python
# find the squeres of list of integers using filter & lambda
arr = [1,2,3,4,5]
print(list(filter(lambda x : x % 2 == 0, arr)))
```

```python
string = input()
c = {}
for char in string:
    if char in c:
        c[char] +=1
    else:
        c[char] =1
print(c)
        
```

```python
# find the frequency of each character
string = input()
c = {}
for char in string:
    c[char] = c.get(char,0) + 1
print(c)
        
```
