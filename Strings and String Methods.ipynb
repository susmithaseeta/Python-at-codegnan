## String Methods
- Case Convention Methods
   - upper(): It converts entire string upper case
   - lower(): It converts entire string lower case
   - title(): It converts the first letter of every word into upper and remaining lower
   - swapcase(): It converts Upper case lettter into lower case and Lower to Upper
   - capitalize(): It converts First character into upper case and remaining into lower case

```python
string = "The sky is blue and Ocean is Blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- Remove white space or pattern at ends of string 
  - strip(): Bye default it removes the white space at both ends of string 
  - lstrip(): It removes the white space at left side of string 
  - rstrip(): It removes the white space at right side of string 


```python
s="     Codegnan     "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@@@Code@@gnan@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
# Split (): It splits the entrire string into list of substring based on specified separator, by default it split at space 
string = "The sky is blue and Ocean is Blue"
print(string.split())
print(string.split("is"))
print(string.split('e'))
```

```python
# 'sep'.join(iterable) : It joins the all iterable elements with separator
words = ['The', 'sky', 'is', 'blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
print("!".join(words))
```

## Substring indexing finding 
  - index(substring): If substring is exists in string, it returns the first occurence of substring position, otherwise 'ValueError'
  - find(substring): If substring is exists in string, it returns the first occurence of substring position, otherwise '-1'


```python
string = "The sky is blue and Ocean is Blue"
print(string.index('is'))
print(string.index('z'))
```

```python
string = "The sky is blue and Ocean is Blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is',10)) # It start checking from 10th position 
print(string.find('e',10))
print(string.find('e',-5))
```

```python
# count(substring): It returns the total non-overlapping occuerence of substring 
s = "ABCBCBCbbaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('z'))
```

```python
# replace(old_substring , new_substring): It replace the old substring with new sustring in string
string = "The sky is blue and Ocean is Blue"
print(string.replace("blue","Red"))
print(string.replace('e','E'))
print(string.replace('e','E',2)) # It replace first 2 occurance
print(string.replace('e','E',3)) # It replace first 3 occurance
```

### String Checking Methods
 **String checking method returns the Boolean value as result**
- isupper(): It returns True, It entire string is Upper case
- islower(): It returns True, It entire string is Lower case
- isaplha(): It returns True, It entire string contains aplhabets
- isdigit(): It returns True, It entire string contains digits
- isalnum(): It returns True, It entire string contains either aplhabet or digits
- startswitch(substring): It returns True, if string starts with substring
- endswitch(substring): It returns True, if string ends with substring

```python
print("CODE".isupper())
print("CODE".islower())
print('code'.islower())
print('1234'.isdigit())
print('1234abc'.isdigit())
print('123abc'.isalpha())
print('Code'.isalpha())
print('Code123'.isalnum())
print("code".startswith('c'))
```
