```python
# characteristics
# ordered collection
t = (1,2,3,[10,30,40],23)
print(t)
print(t[3][2])
```

```python
#immutable
# After operation it doesn't change the length of the object and dosen't change remaining elemnts position
t = (1,2,3,[10,30,40],23)
t[3].append(100)
print(t)
```

```python
# Tuple operations
# Tuple Concatination
# Tuple Repetation
t1 = (1,2,3,4)
t2 = (4,5,6,7,8)
print(t1+t2)
print(t1*2)
print((0,)*10)
```

```python
# Method 
# index(): first occurances
# count(): total occurances
t = (10,20,30,40,50,60,70)
print(t.index(30))
print(t.count(30))
```

```python
# Build-in Function 
# len(), min(), max(), sum()
t = (1,2,3,4,5,6)
print(len(t), min(t),max(t), sum(t))
print("Average:", sum(t) / len(t))
```
