```python
# string to int, float, bool, list, tuple, set
# if string value is number
s = '123456'
s1= ''
print(int(s))
print(float(s))
print(bool(s))
print(bool(s1)) # because it is empty string (s1)
print(list(s))
print(tuple(s))
print(set(s)) # unordered
```

```python
# string to int, float, bool, list, tuple, set
# if string value is float
s = '123456.1205'   #string literal type
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s)) # it present unique values 
```

```python
# list - string, tuple, set
lst = [1, 2, 4.5, 'Hi', 'Hello', 1, 3, 4]
print(str(lst))
print(tuple(lst))
print(set(lst))
```

```python
# Tuple - string, list, set
t = (1, 2, 3, 4,5, 23)
print(str(t))
print(list(t))
print(set(t))
```

```python
# set- string, list, tuple
s = {1, 8, 'a', 2, 3, 2, 3, 22, 20}
print(str(s))
print(list(s))
del tuple
print(tuple(s))
```

```python
# dictionary - string, list, tuple, set
d = {'Name': 'seeta', 'Course': 'python', 'Batch': 57, 'Age': 21}
print(str(d))
print(list(d))
print(set(d))
print(tuple(d))
```
