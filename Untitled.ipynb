```python
units = int(input("Enter units: "))
is_senior = input("Is senior citizen? yes/no: ").lower()

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = units * 2.5
elif units <= 500:
    bill = units * 4
else:
    bill = units * 6
    
if is_senior == "yes":
    bill = bill - (bill * 0.10)

if units > 800:
    bill = bill + (bill * 0.05)

print("Final Bill Amount: Rs", round(bill, 2))
```

```python
#using global scope variable in local scope
x = 100
def outer():
    y = x + 1
    print("Local Scope variable value:",y)
outer()
print("Global scope value:",x) 
```

```python
# Updating the  global scope variable value in local scope
# Can't update global scope variable in local scope directly 
x = 100
def outer():
    x = x + 1
    print("Local Scope variable value:",x)
outer()
print("Global scope value:",x)
```

```python
# Updating the  global scope variable value in local scope by using ' global ' keyword
x = 100
def outer():
    global x
    x = x + 1
    print("Local Scope variable value:",x)
outer()
print("Global scope value:",x)
```

```python
# Updating the Enclosing Scope variable in local scopeby using non-local
x = 100 # global scope variable
def outer():
    x = 200
    x = x + 2
    print("Local scope variable x value:",x)
outer()
print("Global scope variable x value:",x)
```

```python
s1 = "CODE"
s2 = "GNAN"
n = len(s1)
r1 = ""
r2 = ""
for i in range(n):
    if abs(ord(s1[i]) > ord(s2[i])):
        r1 += s1[i].upper()
    else:
        r2 += s2[i].lower()
print(r1)
print(r2)
```

```python


# find bigger and smaller character from s1 and s2
# compare each character in both strins and if s1[i] > s2[i] to bigger and s2[i] to smaller 
# if both same leave it
s1, s2 = map(str,input().upper().split())
n = len(s1)
big = small = ""
for i in range(n):
    if s1[i] != s2[i]:
        if s1[i] > s2[i]:
            big += s1[i]
            small +=s2[i]
        else:
            big +=s2[i]
            small +=s1[i]
print(big)
print(small)
```

```python
# filter the even numbers from list of integers 
def even(x):
    if x % 2 == 0:
        return 1
    return -1
arr = [1,2,3,4,5,6,7]
print(filter(even,arr))
print(list(filter(even,arr)))
print(list(map(even,arr)))
```
