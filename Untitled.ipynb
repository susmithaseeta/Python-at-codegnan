```python
#indexing
#out of range is error 
s="Codegnan"
print(s[3])
print(s[-2])
```

```python
#slicing
# out of range is no error
#str_obj[start:end:stop]
s="Codegnan"
print(s[2:5:1])
print(s[4:7])
print(s[ :5])
print(s[-4: -7:-1])
print(s[5:1:-1])
print(s[7:2])
print(s[5:100])
print(s[10:100])
```

```python
#slicing
#str_obj[start:end:stop]
s="Codegnan"
print(s[2:7])
print(s[4:])
print(s[5:2])
print(s[3:100])
print(s[::-1])
print(s[-3::-1])
print(s[-3::1])
```

```python
# empty string representation
# type() : This Function returns the object data type
s1=""
s2=str()
print(type(s1))
print(type(s2))
```

```python
# String reading from user
#input(): This is function is used to read dynamically from user and this funhction by default read it as string data type 
#dynamic-change,stactic-fixed
s=input()
print(type(s))
print(s)
```
