```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

```python
# 2 dices combination values
res = [(x,y) for x in range (1,7) for y in range(1,7)]
print(res)
```

```python
# return the dice combination if sum of two face value is div by 3
res = [(x,y) for x in range (1,7) for y in range(1,7) if (x +y) % 3 == 0]
print(res)
```

```python
# print any dice whose value is 4 
res = [(x,y) for x in range (1,7) for y in range(1,7) if x == 4 or y == 4]
print(res)
```

```python
# set comprehension 
# 1 to n numbers
n = int(input())
res = {x for x in range (1,n+1)}
print(type(res))
print(res)
```

```python
# dictionary comprehension 
# 1 to n numbers
n = int(input())
res = {x : x ** 2 for x in range (1,n+1)}
print(type(res))
print(res)
```

```python
# tuple comprehension 
# 1 to n numbers
n = int(input())
res = tuple(x for x in range (1,n+1))
print(type(res))
print(res)
```
