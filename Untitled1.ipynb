```python
# retuen 1 to 100 cubic values using function
import sys
import time
s = time.time()
def fun(n):
    res = []
    for i in range(1,n+1):
        res.append(i**3)
        return res
r = fun(1000)
print(sys.getsizeof(r))
print(r)
e = time.time()
print(e-s)
```

```python
# retuen 1 to n values using generator function
import sys
import time
s = time.time()
def gen(n):
    res = []
    for i in range(1,n+1):
        yield i ** 3
r = gen(1000)
print(sys.getsizeof(n))
print(r)
for val in r:
    print(val,end=" ")
e = time.time()
print(e-s)
```
