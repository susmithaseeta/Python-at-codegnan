```python
N = int(input())
numbers = list(map(int, input().split()))

even = 0
odd = 0
positive = 0
negative = 0

for x in numbers:
    if x % 2 == 0:
        even += 1
    else:
        odd += 1

    if x > 0:
        positive += 1
    elif x < 0:
        negative += 1

print("Even:", even)
print("Odd:", odd)
print("Positive:", positive)
print("Negative:", negative)
```

```python
X = int(input())

prime = True

for i in range(2, X):
    if X % i == 0:
        prime = False
        break

if prime:
    print("YES")
else:
    print("NO")
```

```python
N = input()

# Remove leading zeroes after reversing
rev = N[::-1].lstrip('0')

if rev == "":
    rev = "0"

print(rev)

if N == N[::-1]:
    print("YES")
else:
    print("NO")
```

```python
N = int(input())

for num in range(2, N + 1):
    prime = True
    
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    
    if prime:
        print(num, end=" ")
```

```python
 
A, B = map(int, input().split())
gcd = 1

for i in range(1, min(A, B) + 1):
    if A % i == 0 and B % i == 0:
        gcd = i

print(gcd)
```

```python
A, B = map(int, input().split())
result = []

for i in range(A, B + 1):
    s = str(i)
    if all(ch == '4' or ch == '7' for ch in s):
        result.append(i)

if result:
    print(*result)
else:
    print(-1)
```
